```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

# Normal Distribution

![Alt text](gaussian.png)


Note: We can't get the probability of an exact value, we must have a range of values. The exact probability would be zero.

```python
def gaussian(x, mean, std):
    return (1/(np.sqrt(2*np.pi)*std))*np.exp(-((x-mean)**2/(2*std**2)))
```

## Examples:
Assume the average weight of an American adult male is 180 pounds with a standard deviation of 34 pounds. The distribution of weights follows a normal distribution. What is the probability of a weight less than 185?

```python
mean = 180
std = 34
x_val = 185
```

```python
x = np.linspace(0,400,401)
p = gaussian(x, mean, std)
```

```python
plt.plot(x,p)
plt.axvline(x_val, color='red', linestyle='--', label='value')
plt.axvline(mean, color='C0', linestyle='--', label='mean')
plt.axvspan(x_val, 0, alpha=0.25, color='orange')
plt.legend()
plt.show()
```

```python
prob = np.sum(p[x < x_val])
print(f'The probability would be p={prob:.3}')
```

Assume the average weight of an American adult male is 180 pounds with a standard deviation of 34 pounds. The distribution of weights follows a normal distribution. What is the probability of a weight greater or equal than 185?

```python
mean = 180
std = 34
x_val = 185
```

```python
x = np.linspace(0,400,401)
p = gaussian(x, mean, std)
```

```python
plt.plot(x,p)
plt.axvline(x_val, color='red', linestyle='--', label='value')
plt.axvline(mean, color='C0', linestyle='--', label='mean')
plt.axvspan(x_val, 400, alpha=0.25, color='orange')
plt.legend()
plt.show()
```

```python
prob = np.sum(p[x >= x_val])
print(f'The probability would be p={prob:.3}')
```

Assume the average weight of an American adult male is 180 pounds with a standard deviation of 34 pounds. The distribution of weights follows a normal distribution. What is the probability that a man weighs somewhere between 120 and 155 pounds?

```python
mean = 180
std = 34
x_val_min = 120
x_val_max = 155
```

```python
x = np.linspace(0,400,401)
p = gaussian(x, mean, std)
```

```python
plt.plot(x,p)
plt.axvline(x_val_min, color='red', linestyle='--', label='min')
plt.axvline(x_val_max, color='red', linestyle='--', label='max')
plt.axvline(mean, color='C0', linestyle='--', label='mean')
plt.axvspan(x_val_min, x_val_max, alpha=0.25, color='orange')
plt.legend()
plt.show()
```

```python
prob = np.sum(p[(x >= x_val_min)*(x <= x_val_max)])
print(f'The probability would be p={prob:.3}')
```

# Binomial Distribution
![Alt text](binomial.png)

```python
def binomial(n, k, p):
    return (factorial(n)/(factorial(k)*factorial(n-k)))*np.power(p,k)*np.power(1-p,n-k)
```

## Example:
Assume that 15% of the population is allergic to cats. If you randomly select 60 people for a medical trial, what is the probability that 7 of those people are allergic to cats?

```python
n = 60
p = 0.15
k = 7
```

```python
ks = np.arange(0,61)
probs = binomial(n, ks, p)
```

```python
plt.scatter(ks, probs)
plt.axvline(k, color='red', linestyle='--', label='k=7')
plt.title('Binomial distribution')
plt.legend()
plt.grid('on')
plt.show()
```

```python
prob = binomial(60, 7, 0.15)
print(f'The probability would be p={prob:.3}')
```
